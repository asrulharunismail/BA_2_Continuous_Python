```python
import numpy as np
import math

# Problem settings
d = 3
n = 40
nep = 40
MaxIt = 1000

def Sphere(x):
    z = sum(np.power(x, 2))
    return z

def Cost(x):
    return Sphere(x)

def integrated_foraging(x, ass, Vmx, Vmn, sh):
    ngh = np.random.triangular(0, ass, 1)
    Vmx = np.array(Vmx)
    Vmn = np.array(Vmn)
    r = ngh * (Vmx-Vmn) * sh
    nVar = x.size

    y = np.array(x)

    r = ngh * (Vmx-Vmn) * sh
    
    y = y + np.random.uniform(-r, r, size=(1, nVar))

    y = y.flatten()  # Flatten the array

    for i in range (nVar):
        y[i] = max(y[i], Vmn[i])

    return y

class Bee:
    def __init__(self, position, cost):
        self.position = position
        self.cost = cost

# Setting Parameters of Bees Algorithm
varMin = [-10,-10,-10]
varMax = [10,10,10]

maxIt = MaxIt  # Maximum Number of Iterations
shrink = 1

# Initialization
bee = [Bee(np.random.uniform(varMin, varMax, size=(d)), Cost(np.random.uniform(varMin, varMax, size=(d)))) for _ in range(n)]
bee.sort(key=lambda x: x.cost)

BestSol = bee[0]

BestCost = np.zeros(maxIt)
BestPos = [[] for _ in range(maxIt)]

newbee = Bee([], math.inf)

recruitment = np.round(np.linspace(nep, 1, n)).astype(int)

ngh = np.linspace(0, 1, n)

# Main Loop
for it in range(maxIt):
    
    # All Sites
    for i in range(n):

        bestnewbee = Bee([], math.inf)

        assigntment = np.linspace(0, 1, recruitment[i])

        for j in range(recruitment[i]):
            newbee.position = integrated_foraging(bee[i].position, assigntment[j], varMax, varMin, shrink)
            newbee.cost = Cost(newbee.position)
            if np.sum(newbee.cost) < np.sum(bestnewbee.cost):
                bestnewbee.cost = newbee.cost
                bestnewbee.position = newbee.position

        if np.sum(bestnewbee.cost) < np.sum(bee[i].cost):
            bee[i].cost = bestnewbee.cost
            bee[i].position = bestnewbee.position

    shrink = shrink * 0.99

    # Sort
    bee.sort(key=lambda x: np.sum(x.cost))

    # Update
    BestSol.cost = bee[0].cost
    BestSol.position = bee[0].position

    # Store Best Cost ever found
    BestCost[it] = np.sum(BestSol.cost)
    BestPos[it] = BestSol.position
    
    # Display Iteration Information
    print('Iteration {}: Best Cost = {}: Best Position = {}'.format(it, BestCost[it], BestPos[it]))
```
